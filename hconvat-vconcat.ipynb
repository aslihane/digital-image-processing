```python
import numpy as np
import pandas as pd
import cv2
from google.colab.patches import cv2_imshow
```

```python
i1 = cv2.imread("/content/white_cat.png")
cv2_imshow(i1)
```

```python
i2 = cv2.imread("/content/cat.png")
cv2_imshow(i2)
```

```python
resized = cv2.resize(i2, (200,100), interpolation = cv2.INTER_AREA)
cv2_imshow(resized)
```

```python
i3 = cv2.vconcat([i1, i2])
cv2_imshow(i3)
cv2.imwrite('/content/i3.jpg', i3)
```

```python
i4 = cv2.hconcat([i1, i2])
cv2_imshow(i4)
cv2.imwrite('/content/i4.jpg', i4)
```
